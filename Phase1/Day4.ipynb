```python
# 工程思维
# 1. 函数化
# 2. 容错(status_code)
# 3. 通用性
def fetch_post(post_id): # 获取指定post的内容
    url = f"https://jsonplaceholder.typicode.com/posts/{post_id}"
    resp = requests.get(url)
    if resp.status_code != 200:
        print(f"Error: Failed to fetch post {post_id}, status {resp.status_code}")
        return None
    return resp.json()

def get_titles(limit=5): # 获取前limit条标题（不能使用单条postid）
    resp = requests.get("https://jsonplaceholder.typicode.com/posts")
    if resp.status_code != 200:
        print(f"Error: Failed to fetch post {post_id}, status {resp.status_code}")
        return None
    posts = resp.json()
    return [p["title"] for p in posts[:limit]]

post = fetch_post(1)
if post:
    print("单个post title:", post["title"])

limit = 6
titles = get_titles(limit)
print(f"前{limit}条post标题:", titles)
```

```python
import requests
import times
# 超时重试
def safe_get(url, retries=3, timeout=5):
    """
    :param url: 请求URL
    :param retries: 最多尝试次数
    :param timeout: 单次请求超时秒数
    """
    for attempt in range(1, retries+1):
        try:
            resp = requests.get(url, timeout=timeout)
            if resp.status_code == 200:
                return resp.json()
            else:
                print(f"Warning: status_code={resp.status_code} for {url}")
        except requests.exceptions.Timeout:
            print(f"Warning: Timeout on attemp {attempt} for {url}")
        except requests.exceptions.RequestException as e:
            print(f"Warning: Request failed on attempt {attempt}: {e}")

        time.sleep(5)

    # 所有尝试都失败，返回None
    print(f"Error: All {retries} attempts failed for {url}")
    return None
```

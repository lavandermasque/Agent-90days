```python
# Day5最小文件
from fastapi import FastAPI

app = FastAPI()

@app.get("/ping") # HTTP方法为GET，路径为/ping，调用函数ping()
def ping():
    return {"message": "pong"} #return dict

# 终端中启动FastAPI服务
# ./.venv/bin/python -m uvicorn Phase1.main:app --reload
# --reload 改代码自动重启
# summer@danleideMacBook-Pro Agent_90days % ./.venv/bin/python -m uvicorn Phase1.Day5:app --reload

# 验证方式
# 1. 浏览器访问 http://127.0.0.1:8000/ping
# 2. FastAPI自带文档 http://127.0.0.1:8000/docs
```
